```python
# –∏—â–µ–º –∏–Ω–¥–µ–∫—Å –ª—É—á—à–µ–π –æ—Ü–µ–Ω–∫–∏
def find_key(dictionary, value):
    ''' —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –ª—É—á—à–µ–π –æ—Ü–µ–Ω–∫–∏ –≤ –º–∞—Å—Å–∏–≤–µ –∫—Ä–æ—Å—Å –≤–∞–ª–∏–¥–∞—Ü–∏–∏
        
        –∞—Ä–≥—É–º–µ–Ω—Ç—ã:
        dictionary - —Å–ª–æ–≤–∞—Ä—å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –æ–±—É—á–µ–Ω–∏—è
        value - –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –ø–æ–∏—Å–∫–∞
        
        –≤–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ:
        (–∫–ª—é—á –∑–Ω–∞—á–µ–Ω–∏—è, –∏–Ω–¥–µ–∫—Å –∑–Ω–∞—á–µ–Ω–∏—è)
    
    '''
    for key, item in dictionary.items():
        if value in item:
            return key, item.tolist().index(value)
```

```python
rfc_m = find_key(rfc_grid_up.cv_results_, rfc_grid_up.best_score_)
print(rfc_m)
# –Ω–∞—Ö–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ roc_auc –ø–æ –∏–Ω–¥–µ–∫—Å—É
print('ROC_AUC:', rfc_grid_up.cv_results_['mean_test_roc_auc'][rfc_m[1]])
```
